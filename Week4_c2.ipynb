```python
#importing pandas

import pandas
```

```python
# parameters

Revenue_Growth_Factor = 0.18
EBITDA_Margin = 0.45
Taxes = 0.21
Capital_Expenditures = 0.03
working_Capital_Increase = 0.06
EPITDA_Exit_Multiple = 32.70
Discount_Factor = 0.94
```

```python
# showing the data frame for the dates

dates = pandas.date_range(start = '2022', end = '2031', freq = 'Y')
dates
```

```python
# Revenue growth first year

df = pandas.DataFrame(index=dates)
df.loc['2022-12-31', "Revenue"] = 222 * (1 + Revenue_Growth_Factor)
df
```

```python
# Revenue growth throw all years

for i in range (1, len(df)):
    df.iloc[i]["Revenue"] = df.iloc[i-1]["Revenue"] * (1 + 0.18)
df
```

```python
df['EBITDA'] = df['Revenue']*EBITDA_Margin
df["Taxes"] = df ["EBITDA"] * -Taxes
df["Capital Expenditure"] = -df ["Revenue"] * Capital_Expenditures
df["Working Capital"] = -df["Revenue"] * working_Capital_Increase
df["Cash Flow"] = df.drop(columns="Revenue").sum(axis=1)
df
```

```python
df["Discount Factor"] = Discount_Factor
df["Discount Factor"] = df["Discount Factor"].cumprod()
df
```

```python
df["present_value_for_cash_flow"]= df["Discount Factor"] * df["Cash Flow"]
df
```

```python
present_value = df['present_value_for_cash_flow'].sum().round()
present_value
```

```python
terminal_frame = pandas.DataFrame(index=['2032'])
Rev=terminal_frame['Revenue'] = df['Revenue'].iloc[-1] * Revenue_Growth_Factor
EB=terminal_frame['EBITDA'] =terminal_frame['Revenue'] * EBITDA_Margin

Rev= terminal_frame['Revenue'].round()
EB = terminal_frame[['EBITDA']].round()
```

```python
terminal_frame['terminal'] = terminal_frame['EBITDA'] * EPITDA_Exit_Multiple + present_value
Ter_val=terminal_frame['terminal'].round()
```

```python
print(f"""
Our projected Present Value of Projected Future Cash Flows are:${present_value.sum().round(0)}.
We are providing revenue guidance in 2030 to be: ${Rev.values.round()}.
With an estimated EBITDA around: {EB.values}.
The terminal value a this time given a multiple of {EPITDA_Exit_Multiple} and the above EBITDA for 2030, would equal:{Ter_val.values.round()}""")
```
